# [Day 2](https://adventofcode.com/2020/day/2)

```python
txt_file = '/Users/ryanmcgranaghan/Documents/Google_Advent_of_Code_2020/day2_input.txt'
f = open(txt_file,"r")
txt_data = f.read().split('\n')#.remove('')
f.close()

txt_data
```

```python
# Parse into four columns: [low_allowed, high_allowed, letter_to_check, password]
low_allowed = []
high_allowed = []
letter_to_check = []
password = []
for t in range(len(txt_data)):
#     print(txt_data[t])
    tmp = txt_data[t].split(' ')
    low_allowed.append( int( tmp[0].split('-')[0] ) )
    high_allowed.append( int( tmp[0].split('-')[1] ) )
    letter_to_check.append( tmp[1].replace(':','') )
    password.append(tmp[2])
```

```python
cnt_invalid = 0
for i in range(len(txt_data)):
#     print('{}-{} {}: {}'.format(low_allowed[i], high_allowed[i], letter_to_check[i], password[i]))
    tmp_cnt = password[i].count(letter_to_check[i])
    if (tmp_cnt < low_allowed[i]) | (tmp_cnt > high_allowed[i]):
#         print('invalid!')
        cnt_invalid += 1
print( 'the answer to day 2 part 1 = {}'.format(len(txt_data) - cnt_invalid) )
```

```python
cnt_valid = 0
for i in range(len(txt_data)):
#     print('{}-{} {}: {}'.format(low_allowed[i], high_allowed[i], letter_to_check[i], password[i]))
    if ( ( password[i][low_allowed[i]-1] == letter_to_check[i] ) & (password[i][high_allowed[i]-1] != letter_to_check[i]) ) | ( ( password[i][low_allowed[i]-1] != letter_to_check[i] ) & (password[i][high_allowed[i]-1] == letter_to_check[i]) ):
#         print('valid!')
        cnt_valid += 1
print( 'the answer to day 2 part 2 = {}'.format(cnt_valid) )
```

# [Day 3](https://adventofcode.com/2020/day/3)

```python
txt_file = '/Users/ryanmcgranaghan/Documents/Google_Advent_of_Code_2020/day3_input.txt'
f = open(txt_file,"r")
txt_data = f.read().split('\n')
f.close()

txt_data
```

```python
print('rows to traverse = {}'.format(len(txt_data)))
print('length of columns = {}'.format(len(txt_data[0])))
cnt_trees = 0
row_idx = 0
col_idx = 0
while row_idx < (len(txt_data)-1):
    row_idx += 1
    # duplicate the pattern an arbitary amount of times
    chc_row = txt_data[row_idx] * 40
    col_idx += 3
    print(row_idx,col_idx)
    if (chc_row[col_idx] == '#'):
        print('......hit a tree!')
        cnt_trees += 1
        
print( 'the answer to day 3 part 1 = {}'.format(cnt_trees) )
```

```python
def trees_counter(txt_data,slope):
    cnt_trees = 0
    row_idx = 0
    col_idx = 0
    while (row_idx+slope[0]) <= (len(txt_data)-1):
        row_idx += slope[0]
        # duplicate the pattern an arbitary amount of times
        chc_row = txt_data[row_idx] * 80
        col_idx += slope[1]
#         print(row_idx,col_idx)
        if (chc_row[col_idx] == '#'):
#             print('......hit a tree!')
            cnt_trees += 1    
    
    return cnt_trees
```

```python

# Slopes
#     Right 1, down 1.
#     Right 3, down 1. (This is the slope you already checked.)
#     Right 5, down 1.
#     Right 7, down 1.
#     Right 1, down 2.
slopes = np.array([[1,1],
                  [1,3],
                  [1,5],
                  [1,7],
                  [2,1],] 
                 )
cum_cnt_trees = 1
for s in slopes:
    print(s)
    cnt_trees = trees_counter(txt_data,s)
    print('   {}'.format(cnt_trees))
    cum_cnt_trees = cum_cnt_trees * cnt_trees
print( 'the answer to day 3 part 2 = {}'.format(cum_cnt_trees) )
```

```python
ctr_row = 0

valid_entries = ['byr','iyr','eyr','hgt','hcl','ecl','pid','cid']
valid_entries_cheat = ['byr','iyr','eyr','hgt','hcl','ecl','pid']
pp_validity = np.ones(len(txt_data))

ecl_possibilities = ['amb','blu','brn','gry','grn','hzl','oth']

for l in txt_data:
    tmp_txt = l.split()
    d = dict(x.split(':') for x in l.split())
#     print('------------')
#     print(d)
    for v in valid_entries_cheat:
        if not (v in d.keys()):
            pp_validity[ctr_row] = 0
            print('found invalid passport for missing entry at PPID = {}'.format(ctr_row))
            continue
        elif (v == 'byr'):
            if ( int(d.get(v)) < 1920 ) | ( int(d.get(v)) > 2002 ):
                pp_validity[ctr_row] = 0
                print('found invalid passport for byr at PPID = {}'.format(ctr_row))
                continue 
        elif (v == 'iyr'):
            if ( int(d.get(v)) < 2010 ) | ( int(d.get(v)) > 2020 ):
                pp_validity[ctr_row] = 0
                print('found invalid passport for iyr at PPID = {}'.format(ctr_row))
                continue 
        elif (v == 'eyr'):
            if ( int(d.get(v)) < 2020 ) | ( int(d.get(v)) > 2030 ):
                pp_validity[ctr_row] = 0
                print('found invalid passport for eyr at PPID = {}'.format(ctr_row))
                continue 
        elif (v == 'hgt'):
            if not ( d.get(v)[-2:] == 'cm' ) | ( d.get(v)[-2:] == 'in' ):
                pp_validity[ctr_row] = 0
                print('found invalid passport for hgt (general) at PPID = {}'.format(ctr_row))
                continue
            elif ( d.get(v)[-2:] == 'cm' ):
                if ( int(d.get(v)[:-2]) < 150 ) | ( int(d.get(v)[:-2]) > 193 ):
                    pp_validity[ctr_row] = 0
                    print('found invalid passport for hgt centimeters at PPID = {}'.format(ctr_row))
                    continue 
            elif ( d.get(v)[-2:] == 'in' ):
                if ( int(d.get(v)[:-2]) < 59 ) | ( int(d.get(v)[:-2]) > 76 ):
                    pp_validity[ctr_row] = 0
                    print('found invalid passport for hgt inches at PPID = {}'.format(ctr_row))
                    continue     
        elif (v == 'hcl'):
            if not ( d.get(v)[0] == '#' ):
                pp_validity[ctr_row] = 0
                print('found invalid passport for hcl leading # at PPID = {}'.format(ctr_row))
                continue
            else:
                if re.search(r'[^\.0-9a-f]',d.get(v)[1:]):
                    pp_validity[ctr_row] = 0
                    print('found invalid passport for hcl invalid str item PPID = {}'.format(ctr_row))
                    continue
        elif (v == 'ecl'):
            if not any( d.get(v) in ecl_test for ecl_test in ecl_possibilities ):
                pp_validity[ctr_row] = 0
                print('found invalid passport for ecl at PPID = {}'.format(ctr_row))
                continue     
        elif (v == 'pid'):
            if not ( len(d.get(v)) == 9 ):
                pp_validity[ctr_row] = 0
                print('found invalid passport for pid at PPID = {}'.format(ctr_row))
                continue  
        
  
                        
                
    ctr_row += 1
    
print( 'the answer to day 4 part 2 = {}'.format(np.sum(pp_validity)) )

```

```python
txt_file = '/Users/ryanmcgranaghan/Documents/Google_Advent_of_Code_2020/day5_input.txt'
f = open(txt_file,"r")
txt_data = f.read().split('\n')
f.close()

txt_data
```

```python
# input_directions = 'FBFBBFFRLR'


def seat_finder(input_directions):
#     print(input_directions)
    high_end = 127
    low_end = 0
    rng = (high_end - low_end)+1

#     print('{} - {}'.format(low_end,high_end))
    for r_dir in range(7):
#         print(input_directions[r_dir])
        if (input_directions[r_dir] == 'B'):
            high_end = high_end
            low_end = low_end + math.floor(rng/2)
        else:
            high_end = high_end - math.floor(rng/2)
            low_end = low_end
        rng = (high_end - low_end)+1

#         print('{} - {}'.format(low_end,high_end))

    output_row = low_end
#     print('\n\n')
    high_end = 7
    low_end = 0
    rng = (high_end - low_end)+1
    for c_dir in range(3):
#         print(input_directions[7+c_dir])
        if (input_directions[7+c_dir] == 'R'):
            high_end = high_end
            low_end = low_end + math.floor(rng/2)
        else:
            high_end = high_end - math.floor(rng/2)
            low_end = low_end
        rng = (high_end - low_end)+1

#         print('{} - {}'.format(low_end,high_end))


    output_column = low_end

    seat_ID = output_row*8 + output_column
    return seat_ID#[output_row,output_column]
```

```python
seat_ID = []
for seat_chc in txt_data:

    seat_ID.append(seat_finder(str(seat_chc)))
print( 'the answer to day 5 part 1 = {}'.format(np.max(seat_ID)) )
```

```python
seat_ID = np.sort(np.asarray(seat_ID).flatten())
print( 'the answer to day 5 part 2 = {}'.format(seat_ID[np.argwhere( (seat_ID[1:] - seat_ID[:-1]) == 2 )[0][0]]+1) )

```

```python
bags_list = []
contains_list = []
for t in txt_data:
    bags_list.append(t.split('contain')[0])
    contains_list.append(t.split('contain')[1])
```

```python
def search_contains_list(contains_list,bag_id):
    idx = []
    for c in range(len(contains_list)):
        if (bag_id in contains_list[c]):
            idx.append(c)
            
    return idx

def search_bags_list(bags_list,bag_id):
    idx = []
    for c in range(len(bags_list)):
        if (bag_id in bags_list[c]):
            idx.append(c)
            
    return idx
```

```python
ctr = 0
ctr_cum = 0
bag_id_list = ['shiny gold']
bags_super_list = []

while len(bag_id_list) > 0:
    print()
    print('pass through the list = {}'.format(ctr+1))
    
    bag_id_list_tmp = []
    for bag_id in bag_id_list:
        if (bag_id[-1] == 's'):
            bag_id = bag_id[:-1]
        print('....working on bag id = {}'.format(bag_id))
        idx_loop = search_contains_list(contains_list,bag_id)
#         bag_id_list_tmp = list(set( bag_id_list_tmp + [bags_list[b].strip() for b in idx_loop] ))
        bag_id_list_tmp = bag_id_list_tmp + [bags_list[b].strip() for b in idx_loop]
        print('  ------> containing bags found = {}'.format(bag_id_list_tmp))
    ctr_cum += len(set(bag_id_list_tmp))
    print(ctr_cum)
    print(bag_id_list_tmp)
    bags_super_list = bags_super_list + bag_id_list_tmp
    bag_id_list = list(set(bag_id_list_tmp))
#         ctr_bag_id += len(bag_id_list)
    ctr += 1
    print()
    
print('the answer to day 7 part 1 = {}'.format(len(set(bags_super_list))))
```

```python
def id_bags(bag_id):
        
    global cum_sum
    if (bag_id[-1] == 's'):
        bag_id = bag_id[:-1]
    print('working on bag id = {}'.format(bag_id))

    # get all bags contained in this bag
    idx_loop = search_bags_list(bags_list,bag_id)
    children = contains_list[idx_loop[0]].split(',')
    print('   children = {}'.format(children))
    
    if (' no other bags.' in children):
        print('          accessed bottom')
        return 
    
    else: 
        bag_id_list = []
        for i in children:
            bag_id_list.append(i.strip()[2:])
            bag_id_list[-1] = bag_id_list[-1].replace('.','')
    
        multiplier = [int(c.strip()[0]) for c in children]  
        for b in range(len(bag_id_list)):
            print('     b = {}'.format(b))
            print('     bag_id = {}'.format(bag_id_list[b]))
            print('     multiplier = {}'.format(multiplier[b]))
            
            for m in range(multiplier[b]):
                cum_sum += 1
                id_bags(bag_id_list[b])
            print('     cum_sum = {}'.format(cum_sum))
        return cum_sum
        
        
```
